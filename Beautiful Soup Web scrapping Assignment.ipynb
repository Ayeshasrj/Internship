# 2)Write s python program to display list of respected former presidents of India(i.e. Name , Term ofoffice) from {https://presidentofindia.nic.in/former-presidents.htm and make data frame.

```python
page = requests.get('https://presidentofindia.nic.in/former-presidents.htm')
soup = BeautifulSoup(page.content)
idk = soup.find_all('div',class_="presidentListing")
aame = []
lame = []

for i in idk:
    for j in i.find_all('h3'):
        aame.append(j.text.replace('\n',''))
aame

for k in soup.find_all('div',class_="presidentListing"):
    for h in k.find_all('p'):
        lame.append(h.text)
del lame[1:9:2]

df= pd.DataFrame({'Respected President Name':aame, 'Term of office':lame})
df1=df.set_index('Respected President Name')
df1
```

# 3)	Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape and make data frame-
a)	Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
b)	Top 10 ODI Batsmen along with the records of their team andrating.
c)	Top 10 ODI bowlers along with the records of their team andrating.


```python
#a)	Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

page=requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
soup=BeautifulSoup(page.content)
team=[]
mtch1=[]
mtch2=[]
pnt1=[]
pnt2=[]
rating1=[]
rating2=[]

for i in soup.find_all('span',class_="u-hide-phablet"):
    team.append(i.text)

for j in soup.find_all('td',class_="rankings-block__banner--matches"):
    mtch1.append(j.text)

for k in soup.find_all('td',class_="table-body__cell u-center-text"):
    mtch2.append(k.text)
del mtch2[1:39:2]
mtch=mtch1+mtch2

for l in soup.find_all('td',class_="rankings-block__banner--points"):
    pnt1.append(l.text)

for m in soup.find_all('td',class_="table-body__cell u-center-text"):
    pnt2.append(m.text)
del pnt2[0:39:2]
pnt=pnt1+pnt2

for n in soup.find_all('td',class_="rankings-block__banner--rating u-text-right"):
    rating1.append(n.text.replace('\n',''))

for o in soup.find_all('td',class_="table-body__cell u-text-right rating"):
    rating2.append(o.text)
rating=rating1+rating2
import pandas as pd
import numpy as np
df=pd.DataFrame({'Team':team,'Matches':mtch,'Points':pnt,'Rating':rating})
print('Top 10 ODI teams in men’s cricket :')
df.index = np.arange(1,len(df)+1)
df.head(10)
```

```python
#b)	Top 10 ODI Batsmen along with the records of their team andrating.
page=requests.get('https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting')
soup=BeautifulSoup(page.content)
player1=[]
player2=[]
team1=[]
team2=[]
rating1=[]
rating2=[]

for a in soup.find_all('div',class_="rankings-block__banner--name-large"):
    player1.append(a.text)

for b in soup.find_all('td',class_="table-body__cell rankings-table__name name"):
    player2.append(b.text.replace('\n',''))
player=player1+player2

for c in soup.find_all('div',class_="rankings-block__banner--nationality"):
    team1.append(c.text.replace('\n',''))

for d in soup.find_all('span',class_="table-body__logo-text"):
    team2.append(d.text)
team=team1+team2

for e in soup.find_all('div',class_="rankings-block__banner--rating"):
    rating1.append(e.text)

for f in soup.find_all('td',class_="table-body__cell rating"):
    rating2.append(f.text)
rating=rating1+rating2
import pandas as pd
df=pd.DataFrame({'Batsmen':player,'Team':team,'Rating':rating})
print('Top 10 ODI Batsmen :')
df.index = np.arange(1,len(df)+1)
df.head(10)
```

# 5)	Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and make data frame-
i)	Headline
ii)	Time
iii)News Link


```python
page = requests.get('https://www.cnbc.com/world/?region=world')
soup=BeautifulSoup(page.content)
head = []
time = []
link = []

for i in soup.find_all('a',class_="LatestNews-headline"):
    head.append(i.text)
    
for j in soup.find_all('time',class_="LatestNews-timestamp"):
    time.append(j.text)

for k in soup.find_all('a',class_="LatestNews-headline"):
    link.append(k['href'])

import pandas as pd
df = pd.DataFrame({'Headline':head, 'Time':time, 'News Link': link})
df1=df.set_index('Headline')
df1
```

# 7) Write a python program to scrape mentioned details from dineout.co.in and make data frame-
i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

def html_content(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.content)
    
    Hotel_name = []
    Location = []
    for i in soup.find_all('div', class_ = 'restnt-info cursor'):
        Hotel_name.append(i.a.text)
        Location.append(i.div.text)

    Cuisine = []
    for i in soup.find_all('span', class_ = 'double-line-ellipsis'):
        head, sep, tail = i.text.partition('|')
        Cuisine.append(tail)

    link = []
    for i in soup.find_all('img', class_ = 'no-img'):
        link.append(i['data-src'])

    Rating = []
    for i in soup.find_all('div', class_ = 'restnt-rating rating-4'):
        Rating.append(i.text)

    data = pd.DataFrame({'Name of the Hotel':Hotel_name, 'Cuisine of the Hotel':Cuisine, 'Location of the Hotel':Location, 'Rating of the Hotel':Rating, 'Link of the Hotel':link})
    return(data)       
html_content('https://www.dineout.co.in/delhi-restaurants/welcome-back')
```
