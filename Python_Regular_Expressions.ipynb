```python
# 1)Write a Python program to check that a string contains only a certain set of characters (in this case a-z, A-Z and 0-9).

import re

def check_string_characters(string):
    pattern = r'^[a-zA-Z0-9]+$'  # Regular expression pattern for the desired character set
    if re.match(pattern, string):
        print("Matched Successfully")
    else:
        print("String dont match")

# Example usage:
input_string = input("Enter a string: ")
check_string_characters(input_string)
```

```python
#10)Write a Python program that matches a word at the beginning of a string.

import re

def match_word_at_beginning(string, word):
    pattern = r'\b' + re.escape(word)
    if re.match(pattern, string):
        print("The given string starts with the word.")
    else:
        print("The given String does not start with the word.")

# Example usage:
input_string = input("Enter a string: ")
input_word = input("Enter a word: ")
match_word_at_beginning(input_string, input_word)
```

```python
#13 Write a Python program to remove leading zeros from an IP address

def remove_leading_zeros(ip_address):
    parts = ip_address.split(".")
    formatted_parts = [str(int(part)) for part in parts]
    formatted_ip = ".".join(formatted_parts)
    return formatted_ip

# Example usage:
input_ip = input("Enter an IP address: ")
result = remove_leading_zeros(input_ip)
print("Formatted IP address:", result)
```

```python
"""
14)Write a regular expression in python to match a date string in the form of Month name followed by day number and year stored in a text file.
Sample text : ' On August 15th 1947 that India was declared independent from British colonialism, and the reins of control were handed over to the leaders of the Country’.
Output- August 15th 1947
Hint- Use re.match() method here
"""

import re

text = "On August 15th 1947 that India was declared independent from British colonialism, and the reins of control were handed over to the leaders of the Country."

pattern = r'\b[A-Z][a-z]+\s+\d{1,2}(?:st|nd|rd|th)?\s+\d{4}\b'
match = re.match(pattern, text)

if match:
    print("Matched Date String:", match.group())
else:
    print("No Date String Matched")
```

```python
"""
15)Write a Python program to search some literals strings in a string. Go to the editor
Sample text : 'The quick brown fox jumps over the lazy dog.'
Searched words : 'fox', 'dog', 'horse'
"""

def search_literals_in_text(text, searched_words):
    for word in searched_words:
        if word in text:
            print(f"'{word}' found in the text.")
        else:
            print(f"'{word}' not found in the text.")

# Example usage:
sample_text = 'The quick brown fox jumps over the lazy dog.'
searched_words = ['fox', 'dog', 'horse']

search_literals_in_text(sample_text, searched_words)

```

```python
"""
16)Write a Python program to search a literals string in a string and also find the location within the original string where the pattern occurs
Sample text : 'The quick brown fox jumps over the lazy dog.'
Searched words : 'fox'

"""

def search_string(text, pattern):
    locations = []
    text_lower = text.lower()
    pattern_lower = pattern.lower()
    index = text_lower.find(pattern_lower)
    while index != -1:
        locations.append(index)
        index = text_lower.find(pattern_lower, index + 1)
    return locations

# Sample text
text = 'The quick brown fox jumps over the lazy dog.'
pattern = 'fox'

# Searching the pattern in the text
result = search_string(text, pattern)

# Displaying the result
if result:
    print(f"The pattern '{pattern}' was found at the following location(s):")
    for location in result:
        print(f"Index {location}: '{text[location:location+len(pattern)]}'")
else:
    print(f"The pattern '{pattern}' was not found in the text.")
```

```python
""""
17) Write a Python program to find the substrings within a string.
Sample text : 'Python exercises, PHP exercises, C# exercises'
Pattern : 'exercises'.
""""
def find_substrings(text, pattern):
    substrings = []
    index = text.lower().find(pattern.lower())
    while index != -1:
        substrings.append(text[index:index + len(pattern)])
        index = text.lower().find(pattern.lower(), index + 1)
    return substrings

# Sample text
text = 'Python exercises, PHP exercises, C# exercises'
pattern = 'exercises'

# Finding the substrings
result = find_substrings(text, pattern)

# Displaying the result
if result:
    print(f"The pattern '{pattern}' was found in the following substrings:")
    for substring in result:
        print(substring)
else:
    print(f"The pattern '{pattern}' was not found in the text.")
```

```python
#18)Write a Python program to find the occurrence and position of the substrings within a string.

def find_substring_occurrences(text, pattern):
    occurrences = []
    index = text.lower().find(pattern.lower())
    while index != -1:
        occurrences.append((index, text[index:index + len(pattern)]))
        index = text.lower().find(pattern.lower(), index + 1)
    return occurrences

# Sample text
text = 'Python exercises, PHP exercises, C# exercises'
pattern = 'exercises'

# Finding the occurrences and positions of substrings
result = find_substring_occurrences(text, pattern)

# Displaying the result
if result:
    print(f"The pattern '{pattern}' was found at the following positions:")
    for occurrence in result:
        position = occurrence[0]
        substring = occurrence[1]
        print(f"Position: {position} | Substring: {substring}")
else:
    print(f"The pattern '{pattern}' was not found in the text.")
```

```python
#19)Write a Python program to convert a date of yyyy-mm-dd format to dd-mm-yyyy format.

def convert_date_format(date):
    parts = date.split('-')
    if len(parts) == 3:
        year = parts[0]
        month = parts[1]
        day = parts[2]
        return f"{day}-{month}-{year}"
    else:
        return "Invalid date format."

# Example date
date = "2023-07-16"

# Converting the date format
converted_date = convert_date_format(date)

# Displaying the result
print(f"Original date: {date}")
print(f"Converted date: {converted_date}")
```

```python
#20)Write a Python program to find all words starting with 'a' or 'e' in a given string.

import re

def find_words_starting_with_a_or_e(text):
    pattern = r'\b[aAeE]\w+\b'
    matches = re.findall(pattern, text)
    return matches

# Example text
text = "An apple a day keeps the doctor away. Eating an egg every morning is healthy."

# Finding words starting with 'a' or 'e'
result = find_words_starting_with_a_or_e(text)

# Displaying the result
if result:
    print(f"Words starting with 'a' or 'e':")
    for word in result:
        print(word)
else:
    print("No words starting with 'a' or 'e' found.")
```

```python
#21)Write a Python program to separate and print the numbers and their position of a given string.

import re

def separate_numbers(text):
    pattern = r'\d+'
    matches = re.finditer(pattern, text)
    result = []
    for match in matches:
        number = match.group()
        position = match.start()
        result.append((number, position))
    return result

# Example text
text = "I have 10 apples and 5 oranges. The total is 15."

# Separating numbers and their positions
result = separate_numbers(text)

# Displaying the result
if result:
    print("Numbers and their positions:")
    for number, position in result:
        print(f"Number: {number} | Position: {position}")
else:
    print("No numbers found in the text.")
```

```python
#22)Write a regular expression in python program to extract maximum numeric value from a string


import re

def extract_max_numeric_value(text):
    pattern = r'\d+'
    numbers = re.findall(pattern, text)
    if numbers:
        max_number = max(numbers, key=int)
        return int(max_number)
    else:
        return None

# Example text
text = "The numbers are 42, 7, 105, and 19."

# Extracting the maximum numeric value
max_value = extract_max_numeric_value(text)

# Displaying the result
if max_value is not None:
    print(f"The maximum numeric value is: {max_value}")
else:
    print("No numeric values found in the text.")
```

```python
#23)Write a Regex in Python to put spaces between words starting with capital letters

import re

def put_spaces_between_capital_words(text):
    pattern = r'(?<!\s)(?=[A-Z])'
    modified_text = re.sub(pattern, ' ', text)
    return modified_text

# Example text
text = "YouareAmazing."

# Putting spaces between words starting with capital letters
modified_text = put_spaces_between_capital_words(text)

# Displaying the result
print(f"Modified text: {modified_text}")
```

```python
#25)Write a Python program to remove duplicate words from Sentence using Regular Expression
import re

def remove_duplicate_words(sentence):
    pattern = r'\b(\w+)\b(?:\s+\1\b)+'
    modified_sentence = re.sub(pattern, r'\1', sentence)
    return modified_sentence

# Example sentence
sentence = "The cat cat jumped over the wall wall."

# Removing duplicate words
modified_sentence = remove_duplicate_words(sentence)

# Displaying the result
print(f"Modified sentence: {modified_sentence}")
```

```python
#26)Write a python program using RegEx to accept string ending with alphanumeric character.

import re

def is_string_ending_with_alphanumeric(string):
    pattern = r'^.*[a-zA-Z0-9]$'
    match = re.match(pattern, string)
    return bool(match)

# Example strings
strings = ["Hello123", "Testing!", "Python1234", "123456", "Ends with space "]

# Checking if strings end with alphanumeric character
for string in strings:
    if is_string_ending_with_alphanumeric(string):
        print(f"'{string}' ends with an alphanumeric character.")
    else:
        print(f"'{string}' does not end with an alphanumeric character.")
```

```python
"""
27)Write a python program using RegEx to extract the hashtags.
Sample Text: text = RT @kapil_kausik: Doltiwal I mean xyzabc is "hurt" by Demonetization as the same has rendered USELESS <ed><U+00A0><U+00BD><ed><U+00B1><U+0089> "acquired funds" No wo
Output: ['#Doltiwal', '#xyzabc', '#Demonetization']
"""




import re

def extract_hashtags(text):
    pattern = r'#\w+'
    hashtags = re.findall(pattern, text)
    return hashtags

# Sample text
text = """RT @kapil_kausik: #Doltiwal I mean #xyzabc is "hurt" by #Demonetization as the same has rendered USELESS <ed><U+00A0><U+00BD><ed><U+00B1><U+0089> "acquired funds" No wo"""

# Extracting hashtags
hashtags = extract_hashtags(text)

# Displaying the result
print(hashtags)
```

```python
"""

28)Write a python program using RegEx to remove <U+..> like symbols
Check the below sample text, there are strange symbols something of the sort <U+..> all over the place. You need to come up with a general Regex expression that will cover all such symbols.
Sample Text: "@Jags123456 Bharat band on 28??<ed><U+00A0><U+00BD><ed><U+00B8><U+0082>Those who  are protesting #demonetization  are all different party leaders"
Output: @Jags123456 Bharat band on 28??<ed><ed>Those who  are protesting #demonetization  are all different party leaders

"""


import re

def remove_U_plus_symbols(text):
    pattern = r'<U\+[A-F0-9]{4}>'
    modified_text = re.sub(pattern, '', text)
    return modified_text

# Sample text
text = "@Jags123456 Bharat band on 28??<ed><U+00A0><U+00BD><ed><U+00B8><U+0082>Those who are protesting #demonetization are all different party leaders"

# Removing <U+..> symbols
modified_text = remove_U_plus_symbols(text)

# Displaying the result
print(modified_text)

```

```python
'''29)- Write a python program to extract dates from the text stored in the text file.
Sample Text: Ron was born on 12-09-1992 and he was admitted to school 15-12-1999.
Store this sample text in the file and then extract dates.

'''

import re
import urllib.request

def extract_dates_from_text_file(url):
    response = urllib.request.urlopen(url)
    text = response.read().decode('utf-8')
    pattern = r'\d{2}-\d{2}-\d{4}'
    dates = re.findall(pattern, text)
    return dates

# URL of the text file
url = 'https://raw.githubusercontent.com/Ayeshasrj/Internship/main/sample_text.txt'

# Extracting dates from the text file
dates = extract_dates_from_text_file(url)

# Displaying the result
if dates:
    print("Dates extracted from the text file:")
    for date in dates:
        print(date)
else:
    print("No dates found in the text file.")

```

```python
"""
30)Write a Python program to replace all occurrences of a space, comma, or dot with a colon.
Sample Text- 'Python Exercises, PHP exercises.'
Output: Python:Exercises::PHP:exercises:
"""

def replace_space_comma_dot_with_colon(text):
    replacements = {' ': ':', ',': ':', '.': ':'}
    for old_char, new_char in replacements.items():
        text = text.replace(old_char, new_char)
    return text

# Sample text
text = 'Python Exercises, PHP exercises.'

# Replacing space, comma, and dot with colon
modified_text = replace_space_comma_dot_with_colon(text)

# Displaying the result
print(f"Modified text: {modified_text}")

```
